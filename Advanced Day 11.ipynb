```python
new_records = spark.table("events_delta").limit(50)
```

```python
new_records.createOrReplaceTempView("temp_new")

spark.sql("""
INSERT INTO events_delta
SELECT * FROM temp_new
""")
```

```python
spark.sql("DESCRIBE HISTORY events_delta").display()
```

```python
spark.sql("""
SELECT *
FROM events_delta VERSION AS OF 8
LIMIT 20
""").display()
```

```python
current_count = spark.sql("""
SELECT COUNT(*) FROM events_delta
""").collect()[0][0]

old_count = spark.sql("""
SELECT COUNT(*) FROM events_delta VERSION AS OF 8
""").collect()[0][0]

print("Current Count:", current_count)
print("Old Count:", old_count)
```

```python
spark.sql("""
SELECT *
FROM events_delta
EXCEPT
SELECT *
FROM events_delta VERSION AS OF 8
""").display()
```
