```python
events = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv")
)

print(f"Rows loaded: {events.count():,}")
events.show(5)
```

```python
events
```

```python
events.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("events_delta")
```

```python
for i in range(3):
    events.limit(500) \
        .write.format("delta") \
        .mode("append") \
        .saveAsTable("events_delta")
```
