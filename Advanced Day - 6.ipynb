```python
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator

train = spark.table("train_dataset")
test = spark.table("test_dataset")

evaluator = BinaryClassificationEvaluator(
    labelCol="label",
    metricName="areaUnderROC"
)
```

```python
lr = LogisticRegression(
    featuresCol="features",
    labelCol="label"
)

lr_model = lr.fit(train)

lr_predictions = lr_model.transform(test)

lr_auc = evaluator.evaluate(lr_predictions)

print("Logistic Regression AUC:", lr_auc)
```

```python
rf = RandomForestClassifier(
    featuresCol="features",
    labelCol="label",
    numTrees=50,
    maxDepth=5
)

rf_model = rf.fit(train)

rf_predictions = rf_model.transform(test)

rf_auc = evaluator.evaluate(rf_predictions)

print("Random Forest AUC:", rf_auc)
```

```python
results = []

for trees in [20, 50, 100]:
    for depth in [5, 10]:
        
        rf = RandomForestClassifier(
            featuresCol="features",
            labelCol="label",
            numTrees=trees,
            maxDepth=depth
        )
        
        model = rf.fit(train)
        predictions = model.transform(test)
        auc = evaluator.evaluate(predictions)
        
        results.append((trees, depth, auc))
        
        print(f"Trees: {trees}, Depth: {depth}, AUC: {auc}")
```

```python
best_model = max(results, key=lambda x: x[2])

print("Best Parameters:")
print("NumTrees:", best_model[0])
print("MaxDepth:", best_model[1])
print("Best AUC:", best_model[2])
```

```python
print("Logistic Regression AUC:", lr_auc)
print("Random Forest AUC:", rf_auc)
print("Best Tuned RF AUC:", best_model[2])
```
