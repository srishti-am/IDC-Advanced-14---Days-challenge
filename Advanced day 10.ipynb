```python
import time

start = time.time()

df = spark.sql("""
SELECT *
FROM events_delta
WHERE event_type = 'purchase'
""")

df.count()

end = time.time()

print("Execution Time (No Cache):", end - start)
```

```python
df.explain(True)
```

```python
spark.sql("""
SELECT *
FROM events_delta
WHERE event_type = 'purchase'
""").explain(True)
```

```python
spark.sql("OPTIMIZE events_delta")
```

```python
start = time.time()

spark.sql("""
SELECT *
FROM events_delta
WHERE event_type = 'purchase'
""").count()

end = time.time()

print("Execution Time (After Optimize):", end - start)
```
